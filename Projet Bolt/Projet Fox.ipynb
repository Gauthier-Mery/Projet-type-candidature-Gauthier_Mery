```python
#Pour remplir les NaN de la colonne avec des valeurs logiques présentes dans la colonnes 
df_transactions["synthetic_gender"] = df_transactions["synthetic_gender"].interpolate()

#On vérifie qu'il n y a plus de NaN dans la colonne 
df_transactions["synthetic_gender"].isna().sum()

#On remplace les valeurs 1 et 2 par M et F
df_transactions["synthetic_gender"] = df_transactions["synthetic_gender"].replace({1 : "M", 2: "F"})
```

```python
df_transactions["synthetic_age_generation"] = df_transactions["synthetic_age_generation"].interpolate() #on utilise la fonction d'interpolation pour remplacer les nans dans la colonne 
```

```python
df_transactions["synthetic_age_generation"].unique()
```

```python
import pandas as pd
from datetime import datetime

# Création d'un exemple de DataFrame
data = df_transactions['synthetic_age_generation']
df = pd.DataFrame(data)

# Extraire la portion [1980-1994] avec une expression régulière
df_transactions['age_range'] = df_transactions['synthetic_age_generation'].str.extract(r'\[([^\]]+)\]')

# Fill NaN values with the mode of the 'age_range' column
df_transactions['age_range'].fillna(df_transactions['age_range'].mode()[0], inplace=True)

# Fill NaN values with a default value (0 in this case)
df_transactions['age_range'].fillna('0-0', inplace=True)

# Convertir la chaîne de caractères en deux années
df_transactions[['start_year', 'end_year']] = df_transactions['age_range'].str.extract(r'(\d{4})-(\d{4})').astype(int)

# Calculer les âges en utilisant la soustraction
current_year = datetime.now().year
df_transactions['average_age'] = (current_year - df_transactions['end_year']).astype(str) + '-' + (current_year - df_transactions['start_year']).astype(str)  

df_transactions.drop(['synthetic_age_generation', 'age_range', 'start_year', 'end_year'], axis=1, inplace=True)

#test pour mettre au propre les noms des restaurants mais echec 
```

```python
df_transactions['date'] = df_transactions['order_date'] + ' ' +  df_transactions['order_time'] #création d'une colonne regroupant la date et l'heure de la commande
```
