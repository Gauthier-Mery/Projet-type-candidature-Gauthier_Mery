```python
print(df_2.describe())
```

```python
# Calcul des valeurs minimales de la colonne "Mini achat DIPRO" pour chaque groupe de "Gencod"
df_inter = df_2.groupby("Gencod")["Mini achat DIPRO"].agg(["min"])  
# Le résultat est un DataFrame où chaque ligne correspond à un "Gencod" avec la valeur minimale dans une colonne nommée "min"

# Calcul des valeurs minimales de la colonne "PV MARKETPL CONCURRENTE" pour chaque groupe de "Gencod"
df_externe = df_2.groupby("Gencod")["PV MARKETPL CONCURRENTE"].agg(["min"])
# Le résultat est un autre DataFrame où chaque ligne correspond à un "Gencod" avec la valeur minimale dans une colonne nommée "min"
```

```python
# Fusion des deux DataFrames, df_inter et df_externe, en utilisant la colonne "Gencod" comme clé de jointure
df_final = df_inter.merge(df_externe, on="Gencod", how="left")
# La jointure est de type "left", ce qui signifie que toutes les lignes de df_inter seront conservées,
# même si aucune correspondance n'est trouvée dans df_externe.
# Si une correspondance existe, les colonnes correspondantes de df_externe seront ajoutées à df_final.
```

```python
# Création d'une nouvelle colonne "Prix interne" dans df_final
# Cette colonne copie les valeurs de la colonne "min_x" (valeurs minimales de df_inter)
df_final["Prix interne"] = df_final["min_x"]

# Création d'une nouvelle colonne "Prix externe" dans df_final
# Cette colonne copie les valeurs de la colonne "min_y" (valeurs minimales de df_externe)
df_final["Prix externe"] = df_final["min_y"]
```

```python
# Suppression de la colonne "min_x" du DataFrame df_final
# Cette colonne correspondait aux valeurs minimales de df_inter, maintenant renommées en "Prix interne"
df_final = df_final.drop('min_x', axis=1)

# Suppression de la colonne "min_y" du DataFrame df_final
# Cette colonne correspondait aux valeurs minimales de df_externe, maintenant renommées en "Prix externe"
df_final = df_final.drop('min_y', axis=1)
```

```python
df_final
```

```python
# Calcul du pourcentage de différence entre "Prix interne" et "Prix externe"
df_final['Pourcentage_diff'] = ((df_final['Prix interne'] - df_final['Prix externe']) / df_final['Prix interne']) * 100
```

```python
# Tri du DataFrame df_final par la colonne "Pourcentage_diff" en ordre décroissant
df_final = df_final.sort_values(by='Pourcentage_diff', ascending=False)
```
